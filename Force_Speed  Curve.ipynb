```python
import pandas as pd
import sys
import numpy as np
%matplotlib inline
import pylab
import matplotlib
import matplotlib.pyplot as plt
```

```python
#Change file dir, file name

# No Load Data
no_load = pd.read_csv('Data/no_load/PWM_0.1_Trial_3.csv')
no_load_2 = pd.read_csv('Data/no_load/PWM_0.2_Trial_1.csv')
  
# Load Data
l_data_1 = pd.read_csv('Data/load/PWM_0.1_L1_Trial_2.csv')
l_data_2 = pd.read_csv('Data/load/PWM_0.1_L2_Trial_2.csv')
l_data_3 = pd.read_csv('Data/load/PWM_0.1_L3_Trial_3.csv')
l_data_4 = pd.read_csv('Data/load/PWM_0.1_L4_Trial_2.csv')
l_data_5 = pd.read_csv('Data/load/PWM_0.1_L5_Trial_2.csv')
l_data_6 = pd.read_csv('Data/load/PWM_0.1_L6_Trial_2.csv')

# Load Data
l2_data_6 = pd.read_csv('Data/load/PWM_0.2_L6_Trial_1.csv')
l2_data_7 = pd.read_csv('Data/load/PWM_0.2_L7_Trial_3.csv')
```

```python
def st_t_onumber(x):
    import numbers
    # if any number
    if isinstance(x,numbers.Number):
        return x
    # if non a number try convert string to float or it
    for type_ in (int, float):
        try:
            return type_(x)
        except ValueError:
            continue
```

```python
no_load_data = [];
for i in range(len(no_load)):
    A = no_load['Angular Position,Angular Velocity'][i].split(',')
    A = [st_t_onumber(x) for x in A]
    no_load_data.append(A)
##--------------------------------------------------------------------    
load_data_1 = [];
for i in range(len(l_data_1)):
    A = l_data_1['Angular Position,Angular Velocity'][i].split(',')
    A = [st_t_onumber(x) for x in A]
    load_data_1.append(A)
    
load_data_2 = [];
for i in range(len(l_data_2)):
    A = l_data_2['Angular Position,Angular Velocity'][i].split(',')
    A = [st_t_onumber(x) for x in A]
    load_data_2.append(A)
    
load_data_3 = [];
for i in range(len(l_data_3)):
    A = l_data_3['Angular Position,Angular Velocity'][i].split(',')
    A = [st_t_onumber(x) for x in A]
    load_data_3.append(A)
    
load_data_4 = [];
for i in range(len(l_data_4)):
    A = l_data_4['Angular Position,Angular Velocity'][i].split(',')
    A = [st_t_onumber(x) for x in A]
    load_data_4.append(A)
    
load_data_5 = [];
for i in range(len(l_data_5)):
    A = l_data_5['Angular Position,Angular Velocity'][i].split(',')
    A = [st_t_onumber(x) for x in A]
    load_data_5.append(A)
    
load_data_6 = [];
for i in range(len(l_data_6)):
    A = l_data_6['Angular Position,Angular Velocity'][i].split(',')
    A = [st_t_onumber(x) for x in A]
    load_data_6.append(A)
##--------------------------------------------------------------------        
load_data_6_PWM02 = [];
for i in range(len(l2_data_6)):
    A = l2_data_6['Angular Position,Angular Velocity'][i].split(',')
    A = [st_t_onumber(x) for x in A]
    load_data_6_PWM02.append(A)

load_data_7_PWM02 = [];
for i in range(len(l2_data_7)):
    A = l2_data_7['Angular Position,Angular Velocity'][i].split(',')
    A = [st_t_onumber(x) for x in A]
    load_data_7_PWM02.append(A)   
```

```python
no_load_t = [];      # time
no_load_x1 = [];     # angular discplacement
no_load_x2 = [];     # angular velocity
load_1_t = [];      # time
load_1_x1 = [];     # angular discplacement
load_1_x2 = [];     # angular velocity
load_2_t = [];      # time
load_2_x1 = [];     # angular discplacement
load_2_x2 = [];     # angular velocity
load_3_t = [];      # time
load_3_x1 = [];     # angular discplacement
load_3_x2 = [];     # angular velocity
load_4_t = [];      # time
load_4_x1 = [];     # angular discplacement
load_4_x2 = [];     # angular velocity
load_5_t = [];      # time
load_5_x1 = [];     # angular discplacement
load_5_x2 = [];     # angular velocity
load_6_t = [];      # time
load_6_x1 = [];     # angular discplacement
load_6_x2 = [];     # angular velocity
##--------------------------------------------------------------------  
load_6_PWM02_t = [];      # time
load_6_PWM02_x1 = [];     # angular discplacement
load_6_PWM02_x2 = [];     # angular velocity
load_7_PWM02_t = [];      # time
load_7_PWM02_x1 = [];     # angular discplacement
load_7_PWM02_x2 = [];     # angular velocity

for i in range(len(no_load)):
    no_load_t.append(no_load_data[i][0])
    no_load_x1.append(no_load_data[i][1])
    no_load_x2.append(no_load_data[i][2])
for i in range(len(l_data_1)):
    load_1_t.append(load_data_1[i][0])
    load_1_x1.append(load_data_1[i][1])
    load_1_x2.append(load_data_1[i][2])
for i in range(len(l_data_2)):
    load_2_t.append(load_data_2[i][0])
    load_2_x1.append(load_data_2[i][1])
    load_2_x2.append(load_data_2[i][2])
for i in range(len(l_data_3)):
    load_3_t.append(load_data_3[i][0])
    load_3_x1.append(load_data_3[i][1])
    load_3_x2.append(load_data_3[i][2])
for i in range(len(l_data_4)):
    load_4_t.append(load_data_4[i][0])
    load_4_x1.append(load_data_4[i][1])
    load_4_x2.append(load_data_4[i][2])
for i in range(len(l_data_5)):
    load_5_t.append(load_data_5[i][0])
    load_5_x1.append(load_data_5[i][1])
    load_5_x2.append(load_data_5[i][2])
for i in range(len(l_data_6)):
    load_6_t.append(load_data_6[i][0])
    load_6_x1.append(load_data_6[i][1])
    load_6_x2.append(load_data_6[i][2])
##--------------------------------------------------------------------      
for i in range(len(l2_data_6)):
    load_6_PWM02_t.append(load_data_6_PWM02[i][0])
    load_6_PWM02_x1.append(load_data_6_PWM02[i][1])
    load_6_PWM02_x2.append(load_data_6_PWM02[i][2])
for i in range(len(l2_data_7)):
    load_7_PWM02_t.append(load_data_7_PWM02[i][0])
    load_7_PWM02_x1.append(load_data_7_PWM02[i][1])
    load_7_PWM02_x2.append(load_data_7_PWM02[i][2])    
```

```python
diff_vel=np.diff(no_load_x1)*1000
t_diff = no_load_t[0:len(no_load_t)-1]
diff_vel[diff_vel<-5000] = diff_vel[diff_vel<-5000]+2*np.pi*1000
```

```python
# Extracting steady-state velocity data and finding the mean => ss_load_*
g = 9.81
r = 0.011 # radius of the motor shaft in m
F1 = 2.27* g 
F2 = 3.177* g 
F3 = 3.686* g 
F4 = 4.283* g 
F5 = 6.555* g 
F6 = 7.971* g 
F7 = 10.077* g 
start_data = 1000 #PWM 0.1 arrray size is 2950, OWM 0.2 array size is 2500
end_data = 2300

ss_no_load_x2 = no_load_x2[start_data:end_data]
ss_load_1_x2 = load_1_x2[start_data:end_data]
ss_load_2_x2 = load_2_x2[start_data:end_data]
ss_load_3_x2 = load_3_x2[start_data:end_data]
ss_load_4_x2 = load_4_x2[start_data:end_data]
ss_load_5_x2 = load_5_x2[start_data:end_data]
ss_load_6_x2 = load_6_x2[start_data:end_data]
ss_load_6_PWM02_x2 = load_6_PWM02_x2[start_data:end_data]
ss_load_7_PWM02_x2 = load_7_PWM02_x2[start_data:end_data]

ss_no_load = np.mean(ss_no_load_x2)*r #steady state velocity in m/s
ss_load_1 = np.mean(ss_load_1_x2)*r
ss_load_2 = np.mean(ss_load_2_x2)*r
ss_load_3 = np.mean(ss_load_3_x2)*r
ss_load_4 = np.mean(ss_load_4_x2)*r
ss_load_5 = np.mean(ss_load_5_x2)*r
ss_load_6 = np.mean(ss_load_6_x2)*r
ss_load_PWM02_6 = np.mean(ss_load_6_PWM02_x2)*r
ss_load_PWM02_7 = np.mean(ss_load_7_PWM02_x2)*r 
```

```python
# Force-Speed Curve for 0.1
x1 = np.array([ss_no_load,ss_load_1,ss_load_2,ss_load_3,ss_load_4,ss_load_5,ss_load_6])
y1 = np.array([0,F1,F2,F3,F4,F5,F6])
x2 = np.array([ss_load_PWM02_6,ss_load_PWM02_7])
y2 = np.array([F6,F7])

z1 = np.polyfit(x1,y1,1)
p1 = np.poly1d(z1)
xp1 = np.linspace(0,1,10)
plt.plot(x1,y1,'o',xp1,p1(xp1),'-', label = "PWM(0.1)")

z2 = np.polyfit(x2,y2,1)
p2 = np.poly1d(z2)
xp2 = np.linspace(0,1,10)
plt.plot(x2,y2,'o',xp2,p2(xp2),'-', label = "PWM(0.2)")

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylim(0,300)
plt.xlabel('$\dot{\Theta}$ (m/s)')
plt.ylabel(r'Force (N)')
plt.title('Torque (Nm) vs. Speed (m/s) for PWM(0.1)')

plt.show()



# No Load vs. Loads
pylab.plot(no_load_t,no_load_x2,'b',label=r'No load');
pylab.plot(load_1_t,load_1_x2,label=r'2.27kg (hopper)');
pylab.plot(load_2_t,load_2_x2,label=r'3.177kg (hopper+weight)');
pylab.plot(load_3_t,load_3_x2,label=r'3.686kg (hopper+weight)');
pylab.plot(load_4_t,load_4_x2,label=r'4.283kg (hopper+weight)');
pylab.plot(load_5_t,load_5_x2,label=r'6.555kg (hopper+weight)');
pylab.plot(load_6_t,load_6_x2,label=r'7.971kg (hopper+weight)');
pylab.legend(loc='lower right')
plt.xlabel('Time(ms)')
plt.ylabel(r'$\dot{\Theta}$ (m/s)')
plt.title('Load & No Load Response vs. Time(ms)')
plt.show()

# Motor Angular Velocity vs. Calculated Motor Velocity
pylab.plot(t_diff,diff_vel,label=r'Calculated $\dot{\Theta}$');
pylab.plot(no_load_t,no_load_x2,'r',label=r'Motor $\dot{\Theta}$');
pylab.legend(loc='best')
plt.xlabel('Time(ms)')
plt.ylabel(r'$\dot{\Theta}$ (rad/s)')
plt.title('Motor Angular Velocity (rad/s) vs. Calculated Motor Velocity (rad/s)')
plt.show()
```
